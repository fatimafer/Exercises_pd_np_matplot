## Generar un DataFrame con los datos de los cuatro ficheros.

```python
df2016 = pd.read_csv('emisiones-2016.csv', sep=';')
df2016
```

```python
df2019 = pd.read_csv('emisiones-2019.csv', sep=';')
df2019
```

```python
#Juntamos todos los datos de todos los años en un mismo dataset

dfconta = pd.concat([df2016,df2017,df2018,df2019],join='inner')
dfconta
```

## Filtrar las columnas del DataFrame para quedarse con las columnas ESTACION,MAGNITUD, AÑO, MES y las correspondientes a los días D01, D02, etc.

```python
#Para no tardar tanto, voy a usar el for para filtrar las columnas de los días
columnasfiltro = ['ESTACION','MAGNITUD','ANO','MES']
for i in range(1,32):
    if i < 10:
        columnasfiltro.append("D0"+str(i))
    else:
        columnasfiltro.append("D"+str(i))
```

```python
#Esto te ha creado una lista y ya no hace falta poner manualmente los D01,D02...
#Además te añade los corchetes, para que no tengas que poner el doble corchete
print(columnasfiltro)
```

```python
dfconta[columnasfiltro]
```

## Reestructurar el DataFrame para que los valores de los contaminantes de las columnas de los días aparezcan en una única columna.

```python
#Hacemos un melt para que los días aparezcan en la culumna
dfreorganizado = dfconta[columnasfiltro].melt(id_vars= ['ESTACION','MAGNITUD','ANO','MES'],var_name='dia',value_name='valor')
dfreorganizado
```

## Añadir una columna con la fecha a partir de la concatenación del año, el mes y el día (usar el módulo datetime).

```python
#Creamos nueva columna en la que aparezca el día pero sin la d
dfreorganizado['dia2'] = dfreorganizado['dia'].str.slice(-2)
dfreorganizado
```

```python
#Creamos una nueva columna con las fechas en formato str
dfreorganizado['cadena_fecha'] = dfreorganizado['dia2'] + '-' + dfreorganizado['MES'].astype(str) + '-' + dfreorganizado['ANO'].astype(str)
dfreorganizado
```

```python
#La convertimos a formato fecha
#Cuidado, hay que quitarse fechas no válidas como el 30 de febrero o los 31 de septiembre
dfreorganizado['fecha'] = pd.to_datetime(dfreorganizado['cadena_fecha'], format='%d-%m-%Y', errors='coerce')
dfreorganizado
```

## Eliminar las filas con fechas no válidas (utilizar la función isnat del módulo numpy) y ordenar el DataFrame por estaciones contaminantes y fecha.

```python
#Creamos una condición que sea True cuando un valor no tenga fecha
#Filtramos por todos los valores que sean lo contrario de esa condición usando '~'
#Es decir, vamos a quedarnos solo con los valores en que la condición que hemos puesto sea False
#Los que son False son aquellos valores que sí que son en formato fecha
dffechascorrectas = dfreorganizado[~(np.isnat(dfreorganizado['fecha']))]
dffechascorrectas
```

```python
#Ordenamos los valores por estaciones, contaminantes y fecha
dfordenado = dffechascorrectas.sort_values(['ESTACION','MAGNITUD','fecha'],ascending =True)
dfordenado
```

```python
#Aunque no se pida en el ejercicio, yo en mi opinión quitaría algunas columnas con información duplicada
#Así queda más limpio
dflimpio = dfordenado[['ESTACION', 'MAGNITUD', 'valor', 'fecha']]
dflimpio
```

## Crear una función que reciba una estación, un contaminante y un rango de fechas y devuelva una serie con las emisiones del contaminante dado en la estación y rango de fechas dado.

## Crear una función que devuelva las emisiones medias mensuales de un contaminante y un año dados para todos las estaciones.

```python
#Aquí comentí un error y cree un nuevo df eliminando las columnas y meses sin que el enunciado me lo pidiese
#En lugar de corregir ese error, aprovecho y convierto la fecha a año y mes

#PEDIMOS LOS VALORES AL USUARIO
an = int(input('Introduzca el año '))
magni = int(input('Introduzca una magnitud '))

#CREAMOS LA FUNCION
def filtrofechas(dflimpio, an, magni):
    
    #Aquí hacemos la función que hace lo que pide el enunciado,
    #que es filtrar por mes y estación los valores de una magnitud en un año dados
    filtrado = (
        dflimpio[(dflimpio['MAGNITUD']==magni)&(dflimpio['fecha'].dt.year==an)]
    ).groupby([dflimpio['fecha'].dt.month,'ESTACION']
             ).valor.mean()
    #Luego renombramos a mes porque es queda mejor
    filtrado.index = filtrado.index.rename('MES', level=0)
    return filtrado

#APLICAMOS LA FUNCION
filtrofechas(dflimpio, an, magni)
```

## Crear un función que reciba una estación de medición y devuelva un DataFrame con las medias mensuales de los distintos tipos de contaminantes.

```python
#PEDIMOS LOS VALORES AL USUARIO
esta = int(input('Introduzca una estacion '))

#CREAMOS LA FUNCION
def filtrofechas(dflimpio, esta):
    filtrado = pd.DataFrame(
        dflimpio[(dflimpio['ESTACION']==esta)].
        groupby([dflimpio['fecha'].dt.year,dflimpio['fecha'].dt.month]
                                                      ).valor.mean())
    filtrado.index = filtrado.index.rename('MES', level=0)
    filtrado.index = filtrado.index.rename('AÑO', level=1)
    return filtrado

#APLICAMOS LA FUNCION
filtrofechas(dflimpio, esta)
```
