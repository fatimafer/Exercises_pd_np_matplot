## Generar un DataFrame con los datos de los cuatro ficheros.

```python
#Juntamos todos los datos de todos los a√±os en un mismo dataset

dfconta = pd.concat([df2016,df2017,df2018,df2019],join='inner')
dfconta
```

## Filtrar las columnas del DataFrame para quedarse con las columnas ESTACION,MAGNITUD, A√ëO, MES y las correspondientes a los d√≠as D01, D02, etc.

```python
#Para no tardar tanto, voy a usar el for para filtrar las columnas de los d√≠as
columnasfiltro = ['ESTACION','MAGNITUD','ANO','MES']
for i in range(1,32):
    if i < 10:
        columnasfiltro.append("D0"+str(i))
    else:
        columnasfiltro.append("D"+str(i))
```

```python
#Esto te ha creado una lista y ya no hace falta poner manualmente los D01,D02...
#Adem√°s te a√±ade los corchetes, para que no tengas que poner el doble corchete
print(columnasfiltro)
```

```python
dfconta[columnasfiltro]
```

## Reestructurar el DataFrame para que los valores de los contaminantes de las columnas de los d√≠as aparezcan en una √∫nica columna.

```python
#Hacemos un melt para que los d√≠as aparezcan en la culumna
dfreorganizado = dfconta[columnasfiltro].melt(id_vars= ['ESTACION','MAGNITUD','ANO','MES'],var_name='dia',value_name='valor')
dfreorganizado
```

## A√±adir una columna con la fecha a partir de la concatenaci√≥n del a√±o, el mes y el d√≠a (usar el m√≥dulo datetime).

```python
#Creamos nueva columna en la que aparezca el d√≠a pero sin la d
dfreorganizado['dia2'] = dfreorganizado['dia'].str.slice(-2)
dfreorganizado
```

```python
#Creamos una nueva columna con las fechas en formato str
dfreorganizado['cadena_fecha'] = dfreorganizado['dia2'] + '-' + dfreorganizado['MES'].astype(str) + '-' + dfreorganizado['ANO'].astype(str)
dfreorganizado
```

```python
#La convertimos a formato fecha
#Cuidado, hay que quitarse fechas no v√°lidas como el 30 de febrero o los 31 de septiembre
dfreorganizado['fecha'] = pd.to_datetime(dfreorganizado['cadena_fecha'], format='%d-%m-%Y', errors='coerce')
dfreorganizado
```

## Eliminar las filas con fechas no v√°lidas (utilizar la funci√≥n isnat del m√≥dulo numpy) y ordenar el DataFrame por estaciones contaminantes y fecha.

```python
#Creamos una condici√≥n que sea True cuando un valor no tenga fecha
#Filtramos por todos los valores que sean lo contrario de esa condici√≥n usando '~'
#Es decir, vamos a quedarnos solo con los valores en que la condici√≥n que hemos puesto sea False
#Los que son False son aquellos valores que s√≠ que son en formato fecha
dffechascorrectas = dfreorganizado[~(np.isnat(dfreorganizado['fecha']))]
dffechascorrectas
```

```python
#Ordenamos los valores por estaciones, contaminantes y fecha
dfordenado = dffechascorrectas.sort_values(['ESTACION','MAGNITUD','fecha'],ascending =True)
dfordenado
```

```python
#Aunque no se pida en el ejercicio, yo en mi opini√≥n quitar√≠a algunas columnas con informaci√≥n duplicada
#As√≠ queda m√°s limpio
dflimpio = dfordenado[['ESTACION', 'MAGNITUD', 'valor', 'fecha']]
dflimpio
```

## Crear una funci√≥n que reciba una estaci√≥n, un contaminante y un rango de fechas y devuelva una serie con las emisiones del contaminante dado en la estaci√≥n y rango de fechas dado.

## Crear una funci√≥n que devuelva las emisiones medias mensuales de un contaminante y un a√±o dados para todos las estaciones.

```python
#Aqu√≠ coment√≠ un error y cree un nuevo df eliminando las columnas y meses sin que el enunciado me lo pidiese
#En lugar de corregir ese error, aprovecho y convierto la fecha a a√±o y mes

#PEDIMOS LOS VALORES AL USUARIO
an = int(input('Introduzca el a√±o '))
magni = int(input('Introduzca una magnitud '))

#CREAMOS LA FUNCION
def filtrofechas(dflimpio, an, magni):
    
    #Aqu√≠ hacemos la funci√≥n que hace lo que pide el enunciado,
    #que es filtrar por mes y estaci√≥n los valores de una magnitud en un a√±o dados
    filtrado = (
        dflimpio[(dflimpio['MAGNITUD']==magni)&(dflimpio['fecha'].dt.year==an)]
    ).groupby([dflimpio['fecha'].dt.month,'ESTACION']
             ).valor.mean()
    #Luego renombramos a mes porque es queda mejor
    filtrado.index = filtrado.index.rename('MES', level=0)
    return filtrado

#APLICAMOS LA FUNCION
filtrofechas(dflimpio, an, magni)
```

## Crear un funci√≥n que reciba una estaci√≥n de medici√≥n y devuelva un DataFrame con las medias mensuales de los distintos tipos de contaminantes.

```python
#PEDIMOS LOS VALORES AL USUARIO
esta = int(input('Introduzca una estacion '))

#CREAMOS LA FUNCION
def filtrofechas(dflimpio, esta):
    filtrado = pd.DataFrame(
        dflimpio[(dflimpio['ESTACION']==esta)].
        groupby([dflimpio['fecha'].dt.year,dflimpio['fecha'].dt.month]
                                                      ).valor.mean())
    filtrado.index = filtrado.index.rename('MES', level=0)
    filtrado.index = filtrado.index.rename('A√ëO', level=1)
    return filtrado

#APLICAMOS LA FUNCION
filtrofechas(dflimpio, esta)
```
